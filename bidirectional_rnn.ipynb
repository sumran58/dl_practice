```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,Dense,SimpleRNN,Input,Bidirectional
```

```python
(X_train,y_train),(X_test,y_test)=imdb.load_data(num_words=10000)
X_train=pad_sequences(X_train,maxlen=100,padding='post',truncating='post')
X_test=pad_sequences(X_test,maxlen=100,padding='post',truncating='post')
```

```python
model = Sequential([
    Input(shape=(100,)),
    Embedding(input_dim=10000, output_dim=32),

    SimpleRNN(5),
    Dense(1, activation='sigmoid')
])

model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model = Sequential([
    Input(shape=(100,)),
    Embedding(input_dim=10000, output_dim=32),

    Bidirectional(SimpleRNN(5)),
    Dense(1, activation='sigmoid')
])
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

model.summary()
```
