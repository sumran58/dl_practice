```python
import tensorflow as tf
from tensorflow import keras
from keras.layers import Dense, Dropout, Activation, Flatten, Conv2D,MaxPool2D,AvgPool2D
from keras import Sequential
from keras.datasets import mnist
```

```python
(X_train,y_train),(X_test,y_test)=mnist.load_data()
```

#Lenet Architecture

```python
model=Sequential()

model.add(Conv2D(6,activation='tanh',kernel_size=(5,5),padding='valid',input_shape=(32,32,1))) #thse are the convolution layers with 32 filterts and no padidng
model.add(AvgPool2D(pool_size=(2,2),strides=2,padding='valid')) #this is the maxpool layer
model.add(Conv2D(16,activation='tanh',kernel_size=(3,3),padding='valid'))
model.add(AvgPool2D(pool_size=(2,2),strides=2,padding='valid'))

model.add(Flatten()) #1d me convert kiye
model.add(Dense(120,activation='tanh'))
model.add(Dense(84,activation='tanh')) #this is the fully connected layer
model.add(Dense(10,activation='softmax'))
```
