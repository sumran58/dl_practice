```python
import tensorflow as tf
from tensorflow import keras
from keras.layers import Dense, Dropout, Activation, Flatten, Conv2D
from keras import Sequential
from keras.datasets import mnist
```

```python
(X_train,y_train),(X_test,y_test)=mnist.load_data()
```

```python
model=Sequential()
model.add(Conv2D(32,activation='relu',kernel_size=(3,3),padding='valid',input_shape=(28,28,1))) #thse are the convolution layers with 32 filterts and no padidng
model.add(Conv2D(32,activation='relu',kernel_size=(3,3),padding='valid'))
model.add(Conv2D(32,activation='relu',kernel_size=(3,3),padding='valid'))
model.add(Flatten()) #1d me convert kiye
model.add(Dense(128,activation='relu')) #this is the fully connected layer
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
model=Sequential()
model.add(Conv2D(32,activation='relu',kernel_size=(3,3),padding='same',input_shape=(28,28,1))) #thse are the convolution layers with 32 filterts and no padidng
model.add(Conv2D(32,activation='relu',kernel_size=(3,3),padding='same'))
model.add(Conv2D(32,activation='relu',kernel_size=(3,3),padding='same'))
model.add(Flatten()) #1d me convert kiye
model.add(Dense(128,activation='relu')) #this is the fully connected layer
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

##valid is for no padding and same is for padding

```python
model=Sequential()
model.add(Conv2D(32,activation='relu',kernel_size=(3,3),padding='same',strides=(2,2),input_shape=(28,28,1))) #thse are the convolution layers with 32 filterts and no padidng
model.add(Conv2D(32,activation='relu',kernel_size=(3,3),padding='same',strides=(2,2)))
model.add(Conv2D(32,activation='relu',kernel_size=(3,3),padding='same',strides=(2,2)))
model.add(Flatten()) #1d me convert kiye
model.add(Dense(128,activation='relu')) #this is the fully connected layer
model.add(Dense(10,activation='softmax'))
```
