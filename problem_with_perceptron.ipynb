```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
or_data = pd.DataFrame()
and_data = pd.DataFrame()
xor_data = pd.DataFrame()
```

```python
or_data['input1']=[1,1,0,0]
or_data['input2']=[1,0,1,0]
or_data['ouput']=[1,1,1,0]

and_data['input1']=[1,1,0,0]
and_data['input2']=[1,0,1,0]
and_data['ouput']=[1,0,0,0]

xor_data['input1']=[1,1,0,0]
xor_data['input2']=[1,0,1,0]
xor_data['ouput']=[0,1,1,0]
```

```python
sns.scatterplot(x=and_data['input1'],y=and_data['input2'],hue=and_data['ouput'],s=200)
```

```python
sns.scatterplot(x=or_data['input1'],y=or_data['input2'],hue=and_data['ouput'],s=200)
```

```python
sns.scatterplot(x=xor_data['input1'],y=xor_data['input2'],hue=and_data['ouput'],s=200)
```

```python
from sklearn.linear_model import Perceptron
```

```python
clf1=Perceptron()
clf2=Perceptron()
clf3=Perceptron()
```

```python
clf1.fit(and_data.iloc[:,0:2].values,and_data.iloc[:,-1].values)
clf2.fit(or_data.iloc[:,0:2].values,or_data.iloc[:,-1].values)
clf3.fit(xor_data.iloc[:,0:2].values,xor_data.iloc[:,-1].values)
```

```python
clf1.coef_ #.coef_ means weight that how much feature afect the output here if x1 increase y alsdo increases with 2 and x2 increases means y also increase iwth 2 if the number would come in minus means y decreses
```

```python
clf1.intercept_ #intercept_means bias means whats the value of y when the x=0 of the input =0  we can say as the strating level
```
